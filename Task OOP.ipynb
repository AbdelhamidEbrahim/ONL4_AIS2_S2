```python
    """
    Task1 class checks whether a number is even or odd.

    Attributes:
        number (int): The number to check.
    """
    class task1:
        def __init__(self,num):
            self.number = num
            
        def check_even_odd(self):
            if self.number %2 ==0:
                return "Even"
            else:
                return "Odd"
```

```python
Test=task1(5)
print(Test.check_even_odd())
```
